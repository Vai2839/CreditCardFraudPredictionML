```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample=legit.sample(n=len(fraud),random_state=2)
tobetrained_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
tobetrained_df['Class'].value_counts()
```

```python
x=tobetrained_df.drop('Class',axis=1)
y=tobetrained_df['Class']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression,LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,r2_score
```

```python
model={
    'ls':LinearRegression(),
    'lg':LogisticRegression(),
    'lasso':Lasso(),
    'ridge':Ridge(),
    'dtc':DecisionTreeClassifier(),
    'knc':KNeighborsClassifier(),
    'rfc':RandomForestClassifier()
}
for name,mod in model.items():
    mod.fit(xtrain,ytrain)
    ypred=mod.predict(xtest)
    if (name=='lg'):
         print(f'squared error for {name} : {mean_squared_error(ytest,ypred)} absolute error : {mean_absolute_error(ytest,ypred)} aquracy : {accuracy_score(ytest,ypred)}')
    else:
             print(f'squared error for {name} : {mean_squared_error(ytest,ypred)} absolute error : {mean_absolute_error(ytest,ypred)} score : {r2_score(ytest,ypred)}')

    
```
